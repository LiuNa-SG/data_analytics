## operators

The <mark>or</mark> and <mark>and</mark> Python statements require **truth-values**.

<p>

For pandas, these are considered **ambiguous**, so you should use "bitwise" operations

Logic  | Bit-Wise
-------------------|------------------
```and``` | **&**
```or``` | **|**
```not``` | **~**

code works:
```(python)
df = df[(df['age'] > 20) | (df['score'] < 10)]
```

code not work:
```(python)
# ValueError: The truth value of a Series is ambiguous. Use a.empty, a.bool(), a.item(), a.any() or a.all().
df = df[(df['age'] > 20) and (df['score'] < 10)]
```

alternative way: query
```(python)
df = df[(df['age'] > 20) | (df['score'] < 10)]
```
reason:
* (df['age'] > 20) is a series with T/F/T etc
* (df['score'] < 10) is a series with F/T/F etc

```(python)
# any value in either column is True?
(df.C > 0.25).any() or (df.C < -0.25).any() # True
# All values in either column is True?
(df.C > 0.25).all() or (df.C < -0.25).all() # False
```

````python
# Comparisons:
# Equal:            == # if value is same
# Not Equal:        !=
# Greater Than:     >
# Less Than:        <
# Greater or Equal: >=
# Less or Equal:    <=
# Object Identity:  is # if memory id is same

a = [1,2,3]
b = [1,2,3]
c = a
print(f"list a memory id is {id(a)}, while list b memory is {id(b)}. 'a is b' has result {a is b}")
print(f"list a memory id is {id(a)}, while list c memory is {id(c)}. 'a is c' has result {a is c}")

# False Values:
    # False
    # None
    # Zero of any numeric type
    # Any empty sequence. For example, '', (), [].
    # Any empty mapping. For example, {}.
````

## element-wise vs broadcasting

````python
# element-wise (same size) vs broadcast (diff size)

import pandas as pd
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a + b  # Element-wise addition
print(result)  # Output: [5 7 9]


a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 20, 30])
result = a + b  # Broadcasting: b is broadcasted to match the shape of a
print(result)  # Output: [[11 22 33] [14 25 36]]


df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
result = df + 1 # broadcasting the SCALAR value to all elements in the dataframe
print(result)
````

## re in str function

````python
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo
````

````python
# str.match
import re
searchfor = ['OVAL', 'TRIANGLE']
pat = '|'.join(searchfor)
pattern = re.compile(r"\b" + (pat) + r"\b", re.IGNORECASE)
print(pattern)
ufo['Shape Reported New']= ufo['Shape Reported'].str.match(pattern)
ufo
````

````python
# str.search
import re
searchfor = ['AAA', 'OVA']
pat = '|'.join(searchfor)
print(pat)
ufo['Shape Reported New'] = ufo['Shape Reported'].str.contains(pat)
ufo
````

## eMail Pattern

````python
import re

emails = '''
nanaliu@gmail.com
liu.na@tsinghua.edu.cn
liu-na@my-work.net
'''
# r indicate raw string format
# . is written as \.
email_pattern = re.compile(r'[a-zA-z0-9_.+-]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+')

matches = email_pattern.finditer(emails)

for match in matches:
    print(match)
````

## Groups

````python
# $0 is the overall match
# \1 is a first captured group
# \2 is a second captured group
import pandas as pd
import re

df = pd.DataFrame({'name':['Tan, Susan','Yeo, Alex','Wong, Chris']})

df['name2'] = df['name'].str.replace(r'(\w+), (\w+)', r'\2 \1')
df
````

## Text Splits

````python
import re

# Compile a pattern to match one or more non-word characters
non_word_pattern = re.compile(r'\W+')

# Split a string into words
words = non_word_pattern.split('This is a test, with: various; punctuation!')
print(words) # Output: ['This', 'is', 'a', 'test', 'with', 'various', 'punctuation']
````

````python
import re

# Compile a pattern to match any digit
digits_pattern = re.compile(r'\d')

# Replace all digits in a string with '#'
masked_string = digits_pattern.sub('#', 'My phone number is 123-456-7890.')
print(masked_string) # Output: My phone number is ###-###-####.
````
